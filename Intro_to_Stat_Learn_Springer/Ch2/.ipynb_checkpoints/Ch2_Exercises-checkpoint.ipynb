```python
import pandas as pd
import numpy as np
```

```python
X1 = [0,2,0,0,-1,1]
X2 = [3,0,1,1,0,1]
X3 = [0,0,3,2,1,1]
Y = ['Red', 'Red', 'Red', 'Green', 'Green', 'Red']
data = list(zip(X1, X2, X3, Y))  
df = pd.DataFrame(data, columns = ['X1', 'X2', 'X3', 'Y']) 
df
```

```python
euc= []
for row in df.iterrows():
    r = row[1]
    dist = ((0-r.X1)**2 + (0-r.X2)**2 + (0-r.X3)**2)**(1/2)
    euc.append(dist)
df['Dist_to_0'] = euc
df
```

```python
college = pd.read_csv('../data/College.csv')
college["Unnamed: 0"]
college.rename(index=college["Unnamed: 0"], inplace=True)
college = college.iloc[:,1:]
college
```

```python
college.describe()
```

```python
college.dtypes
```

```python
import seaborn as sns
sns.pairplot(college.iloc[:, :10]);
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-white')

college.boxplot(column=['Outstate'], by='Private')
```

```python
def is_elite(row):
    if row.Top10perc > 50:
        return 'Yes'
    return 'No'

college['Elite'] = college.apply(lambda row: is_elite(row), axis=1)
college['Elite'].describe()
```

```python
college.boxplot(column=['Outstate'], by='Elite');
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot()

for i in range(1, 5):
    ax = fig.add_subplot(2, 2, i)
    ax.hist(college.iloc[:, i+1], 10);
    ax.set_title(college.columns[i+1])
```
