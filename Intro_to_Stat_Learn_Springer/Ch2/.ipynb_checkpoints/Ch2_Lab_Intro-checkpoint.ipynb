<h2> Load Data

```python
auto = pd.read_csv('../data/Auto.csv', na_values = '?').dropna()
auto.info()
```

```python
auto.columns
```

```python
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot()
ax.scatter(auto.cylinders, auto.mpg)
```

<h2> Additional Graphical and Numerical Summaries

```python
cyls = list(set(np.array(auto.cylinders)))
data = []
for val in cyls:
    d = np.array(auto.loc[auto['cylinders'] == val].mpg)
    data.append(d)
fig, ax = plt.subplots()
ax.boxplot(data, positions = cyls);
```

```python
fig, ax = plt.subplots()
ax.hist(auto.mpg, 10);
```
